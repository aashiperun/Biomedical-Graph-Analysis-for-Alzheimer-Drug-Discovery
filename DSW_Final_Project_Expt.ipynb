**EXPERIMENT: Comparison of Predicted Drugs for Alzheimer's Disease using both Datasets**

```python
import pandas as pd

# Load the CSV file
dataset1 = pd.read_csv('/content/drive/MyDrive/Spring2023/DSW/DSW_Project/expt/predict_result_scaled_ensemble_approve.csv')
dataset2 = pd.read_csv('/content/drive/MyDrive/Spring2023/DSW/DSW_Project/expt/predict_result_scaled_TransR_approve.csv')
```

```python
# Merge the two datasets on the 'Drug' column
df = pd.merge(dataset1, dataset2, on='Drug')
df
```

```python
# Select the top 10 drugs based on 'Score_scaled' from each dataset
top_df1 = dataset1.nlargest(10, 'Score_scaled')
top_df2 = dataset2.nlargest(10, 'Score_scaled')

# Merge the two datasets on the 'Drug' column
top_df = pd.merge(top_df1, top_df2, on='Drug', how='inner')

# Print the common drugs and their scores in each dataset
for i in range(len(top_df)):
    print(f"Drug: {top_df['Drug'].iloc[i]}")
    print(f"Score in dataset 1: {top_df['Score_scaled_x'].iloc[i]}")
    print(f"Score in dataset 2: {top_df['Score_scaled_y'].iloc[i]}")
    print("---------")
```

```python
df1 = dataset1
df2 = dataset2
# Find the intersection (common drugs) between two dataframes
common_drugs = pd.merge(df1, df2, how='inner', on='Drug')

# Find the difference (unique drugs) between two dataframes
unique_df1 = df1[~df1.Drug.isin(df2.Drug)]
unique_df2 = df2[~df2.Drug.isin(df1.Drug)]

# Count the number of common and unique drugs
num_common_drugs = len(common_drugs)
num_unique_df1 = len(unique_df1)
num_unique_df2 = len(unique_df2)

print(f"Number of common drugs: {num_common_drugs}")
print(f"Number of unique drugs in dataset 1: {num_unique_df1}")
print(f"Number of unique drugs in dataset 2: {num_unique_df2}")
```
